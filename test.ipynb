```python
import pdfplumber
import re

def preprocess_text(text):
    text = text.replace('\n', ' ')
    text = re.sub(r'[^a-zA-Z0-9\s.]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    text = text.lower()
    return text

def extract_and_preprocess_text(file_path):
    preprocessed_texts = []  # List to store preprocessed text for each page

    with pdfplumber.open(file_path) as pdf:
        num_pages = len(pdf.pages)

        for page_num in range(num_pages):
            page = pdf.pages[page_num]
            text = page.extract_text()
            preprocessed_text = preprocess_text(text)
            
            # Append the preprocessed text to the list
            preprocessed_texts.append(preprocessed_text)

            print(f"Page {page_num + 1}:\n{preprocessed_text}\n")

    return preprocessed_texts  # Return the list of preprocessed texts

# Example usage:
pdf_file_path = 'C:/Users/sriyo/Desktop/jets/NLP Bot/files/ML.pdf'
preprocessed_texts = extract_and_preprocess_text(pdf_file_path)

# Now, preprocessed_texts contains the preprocessed text for each page
```

```python
import spacy

# Load the pre-trained English NER model
nlp = spacy.load("en_core_web_sm")

def perform_ner(text):
    # Process the text using SpaCy NER model
    doc = nlp(text)

    # Extract named entities and return them
    named_entities = [(ent.text, ent.label_) for ent in doc.ents]
    return named_entities

# Example usage:
all_named_entities = []  # List to store named entities for each page

for i, preprocessed_text in enumerate(preprocessed_texts, start=1):
    print(f"Page {i}:")
    page_entities = perform_ner(preprocessed_text)
    all_named_entities.append({"page": i, "entities": page_entities})
    print("\n")

# Now, all_named_entities contains the named entities for each page
# You can save this information to a file or use it as needed
```

```python
def generate_qa_pairs(entities):
    qa_templates = [
        {"question": "What is the significance of {entity}?", "answer": "The significance of {entity} is..."},
        {"question": "When did {entity} happen?", "answer": "{entity} happened in..."},
        {"question": "Explain the role of {entity} in {label}.", "answer": "The role of {entity} in {label} is..."}
    ]

    generated_qa_pairs = []

    for page_info in entities:
        page_num = page_info["page"]
        page_entities = page_info["entities"]

        for entity, label in page_entities:
            for template in qa_templates:
                generated_question = template["question"].format(entity=entity, label=label)
                generated_answer = template["answer"].format(entity=entity, label=label)
                
                generated_qa_pairs.append({"page": page_num, "question": generated_question, "answer": generated_answer})

    return generated_qa_pairs

# Example usage:
identified_entities = all_named_entities
qa_pairs = generate_qa_pairs(identified_entities)

# Print the generated QA pairs
for i, qa_pair in enumerate(qa_pairs, start=1):
    print(f"Question {i} (Page {qa_pair['page']}): {qa_pair['question']}")
    print(f"Answer {i}: {qa_pair['answer']}\n")
```
